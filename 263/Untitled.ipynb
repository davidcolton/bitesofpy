```python
from itertools import product
from collections import defaultdict
from collections import deque


def island_number():
    island_number = 1
    while True:
        yield f'{island_number}'
        island_number += 1

def get_neighbours(position, grid_size):
    row, col = position
    north = (row - 1, col)
    east = (row, col + 1)
    south = (row + 1, col)
    west = (row, col - 1)
    neighbours = [north, east, south, west]
    return [neighbour for neighbour in neighbours
            if 0 <= neighbour[0] < grid_size 
            and 0 <= neighbour[1] < grid_size ]

size_of_grid = len(grid)
grid_positions_to_visit = list(product(range(size_of_grid), range(size_of_grid)))

islands = island_number()
number_of_islands = 0
```

```python

try:    
    land_coordinates = [position 
                        for position in grid_positions_to_visit
                        if grid[position[0]][position[1]] == 1]
except IndexError:
    return number_of_islands
```

```python



while land_coordinates:
    visited = set()
    neighbours = deque()
    visiting = land_coordinates[0]
    visited.add(visiting)
    [neighbours.append(n) for n in get_neighbours(visiting, size_of_grid)]
    while neighbours:
        neighbour = neighbours.popleft()
        if neighbour in land_coordinates and neighbour not in visited:
            visited.add(neighbour)
            [neighbours.append(n) for n in get_neighbours(neighbour, size_of_grid)]
    
    number_of_islands = next(islands)
    [land_coordinates.remove(pos) for pos in visited]
            
return number_of_islands        
    
    



```
